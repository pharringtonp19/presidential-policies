### **Import Libraries**

```python
import os
from fredapi import Fred
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import userdata
from matplotlib import font_manager
import matplotlib.dates as mdates
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams['image.interpolation'] = 'nearest'
rcParams['image.cmap'] = 'viridis'
rcParams['axes.grid'] = False
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

### **FRED API**

```python
fred = Fred(api_key=userdata.get('FRED'))
```

### **Read In Data**

```python
data_series1 = fred.get_series('DGS30').to_frame('30')
data_series2 = fred.get_series('DGS10').to_frame('10')
df = pd.concat([data_series1, data_series2], axis=1)
# Drop missing rows if any
df = df.dropna()
```

```python
x = df.index
y = df['10'].values

# --- NYT-style colors ---
line_color = '#1f77b4'
fill_color = "#e6f1fb"
forecast_fill = "#dfeaf7"
grid_color = '#e6e6e6'

fig, ax = plt.subplots(figsize=(7, 4.5), dpi=300)
ax.set_facecolor("white")
fig.patch.set_facecolor("white")

# area + line
ax.fill_between(x, 0, y, color=fill_color, linewidth=0, zorder=1, alpha=0.6)
ax.yaxis.grid(True, color=grid_color, linewidth=.5, zorder=2)
ax.plot(x, y, color=line_color, linewidth=1, zorder=3,
        solid_capstyle="round", solid_joinstyle="round")
# spines & ticks (keep only a subtle bottom spine)
for side in ("top", "left", "right"):
    ax.spines[side].set_visible(False)
ax.spines["bottom"].set_color("#bdbdbd")
ax.spines["bottom"].set_linewidth(1)

ax.tick_params(axis="x", length=2, labelsize=9, pad=2)
ax.tick_params(axis="y", length=0, labelsize=9, pad=2)


# tidy limits
ax.set_ylim(0, max(y)*1.10)
ax.set_axisbelow(True)            # any future gridlines stay behind

# title + subtitle (left-aligned)
ax.text(0., 1.05, s='Ten-Year U.S. Treasuries', transform=ax.transAxes, size=14)
ax.text(0, 1.01, "Yield", transform=ax.transAxes,
        fontsize=9, color="#555555")

ax.annotate("Source: Board of Governors of the Federal Reserve System (US) via FRED",
        xy = (0., -0.1),
        xycoords='axes fraction',
        ha='left',
        va="center",
        fontsize=8)

last_x = x[-1]
last_y = y[-1]


# annotate it
ax.annotate(f"{last_y:.2f}%",
            xy=(last_x, last_y),             # point to annotate
            xytext=(5, 0),                   # offset text a bit to the right
            textcoords="offset points",
            ha="left", va="center",
            fontsize=8, color=line_color,
            bbox=dict(boxstyle="round,pad=0.2",
                      fc="white", ec="none", alpha=0.7))
yticks = ax.get_yticks()
ax.set_yticks([t for t in yticks if t != 0])
fig.savefig("ten_year_yield.png", bbox_inches="tight", dpi=300)
plt.show()
```
