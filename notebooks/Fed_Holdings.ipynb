### **Import Libraries**

```python
import os
from fredapi import Fred
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import userdata
from matplotlib import font_manager
import matplotlib.dates as mdates
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams['image.interpolation'] = 'nearest'
rcParams['image.cmap'] = 'viridis'
rcParams['axes.grid'] = False
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

### **FRED API**

```python
fred = Fred(api_key=userdata.get('FRED'))
```

```python
# --- Data ---
# Fed holdings of Treasuries (SOMA) – weekly, billions $
treasuries = fred.get_series('WSHOSHO').to_frame('Treasuries')
mbs = fred.get_series('WSHOMCB').to_frame('MBS')
df = pd.merge(treasuries, mbs, left_index=True, right_index=True, how='inner')
df['Treasuries'] = df['Treasuries'] /1e3
df['MBS'] = df['MBS'] /1e3
df.head()
#fed_treasuries = fed_treasuries / 1e3   # convert millions → billions
```

```python
x = df['MBS'].index
y = df['MBS'].values

# --- NYT-style colors ---
line_color = '#1f77b4'
fill_color = "#e6f1fb"
forecast_fill = "#dfeaf7"
grid_color = '#e6e6e6'

fig, ax = plt.subplots(figsize=(7, 4.5), dpi=300)
ax.set_facecolor("white")
fig.patch.set_facecolor("white")

# area + line
ax.fill_between(x, 0, y, color=fill_color, linewidth=0, zorder=1, alpha=0.6)
ax.yaxis.grid(True, color=grid_color, linewidth=.5, zorder=2)
ax.plot(x, y, color=line_color, linewidth=1, zorder=3,
        solid_capstyle="round", solid_joinstyle="round")

ax.plot(df.index, df['Treasuries'], color='purple', linewidth=1, zorder=3,
        solid_capstyle="round", solid_joinstyle="round")

# spines & ticks (keep only a subtle bottom spine)
for side in ("top", "left", "right"):
    ax.spines[side].set_visible(False)
ax.spines["bottom"].set_color("#bdbdbd")
ax.spines["bottom"].set_linewidth(1)

ax.tick_params(axis="x", length=2, labelsize=9, pad=2)
ax.tick_params(axis="y", length=0, labelsize=9, pad=2)


# tidy limits
ax.set_axisbelow(True)            # any future gridlines stay behind

# title + subtitle (left-aligned)
ax.text(0., 1.05, s='Fed Holdings', transform=ax.transAxes, size=14)
ax.text(0, 1.01, "Billions", transform=ax.transAxes,
        fontsize=9, color="#555555")

ax.annotate("Board of Governors of the Federal Reserve System (US) via FRED",
        xy = (0., -0.1),
        xycoords='axes fraction',
        ha='left',
        va="center",
        fontsize=8)

ax.annotate("Treasuries",
        xy = (.92, .8),
        xycoords='axes fraction',
        ha='left',
        va="center",
        weight='bold',
        fontsize=8)

ax.annotate("MBS",
        xy = (.92, .34),
        xycoords='axes fraction',
        ha='left',
        va="center",
        weight='bold',
        fontsize=8)

plt.ticklabel_format(style='plain', axis='y')  # <- disable sci notation on y-axis

fig.savefig("fed_holdings.png", bbox_inches="tight", dpi=300)
plt.show()
```
